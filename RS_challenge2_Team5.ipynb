## Converting pandas dataframe to surprise dataset

```python
reader = Reader(rating_scale=(0,10))
data = Dataset.load_from_df(ratings[['User-ID', 'ISBN', 'Book-Rating']], reader)
```

```python
type(data)
```

```python
#Train test split
train_data, test_data = train_test_split(data, test_size=0.2)
```

## SVD

```python
svd = SVD()
svd.fit(train_data)
```

```python
#Making predictions
pred = svd.test(test_data)
```

```python
pred
```

pred contains the predictions as a list in the form of tuple (user, item, actual_rating, predicted_rating, details)

```python
accuracy.rmse(pred)
```

```python
accuracy.mae(pred)
```

The RMSE value for rating prediction is 3.5033 and MAE is 2.8192.
